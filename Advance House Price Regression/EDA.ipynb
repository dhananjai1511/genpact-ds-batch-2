## Numerical Features

```python
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O']

print("Total numerical features are", len(numerical_features))

df[numerical_features].head()
```

### Analysing Temporal Variables

```python
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median Sale Price')
plt.title("House Price vs year sold")
plt.show()
```

```python
for feature in year_features:
    if feature != 'YrSold':
        data = df.copy()
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
    
```

### Descrete Features

```python
descrete_features = [feature for feature in numerical_features if len(df[feature].unique()) < 25 and feature not in year_features + ['Id']]

print("Total descrete count", len(descrete_features))
```

```python
for f in descrete_features:
    print(f)
```

```python
for feature in descrete_features:
    data = df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

### Continious Variables

```python
continious_features = [feature for feature in numerical_features if feature not in descrete_features + year_features + ['Id']]
```

```python
len(continious_features)
```

```python
continious_features
```

```python
for feature in continious_features:
    df[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    
    plt.show()    
```

### Log Normal Transformation

```python
for feature in continious_features:
    if 0 in df[feature].unique():
        pass
    else:
        
        
        plt.scatter(df[feature], df['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        
        plt.show()        
```

```python
for feature in continious_features:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        
        plt.show()        
```

### Outliers

```python
for feature in continious_features:
    if 0 in data[feature].unique():
        pass
    else:
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
